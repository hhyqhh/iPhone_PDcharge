```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import numpy as np
excel_str = ['anker18w_iphonexr.xlsx',
             'lulian18w_iphonexr.xlsx',
             'xiaomi64w_iphonexr.xlsx',
             'apple10w_iphonexr.xlsx',
             'apple5w_iphonexr.xlsx'
            ]
```

```python
def plot_voltage_current(files,title_str):
    raw_data=pd.read_excel('./data/{}'.format(files)).values
    datas = pd.DataFrame(raw_data[:,1:],columns=['voltage','current'])
    datas['power'] = datas['voltage']*datas['current']
    
    fig = plt.figure(figsize=(8,4))
    ax1 = fig.add_subplot(111)
    ax1.plot(datas.index,datas['current'],'r',label='Current')
    plt.ylim(0,2.5)
    plt.legend(loc = 'upper left')
    ax2 = ax1.twinx()
    ax2.plot(datas.index,datas['voltage'],'b',label='Voltage')

    plt.legend(loc = 'upper right')

    ax1.set_xlabel('Times (min)')
    ax1.set_ylabel('Current (A)', color='r')
    ax2.set_ylabel('Voltage (V)', color='b')
    
    
    if title_str == 'Apple 5w current and voltage curve':
        plt.ylim(0,14)
        plt.xlim(0,1260)

        old_ticks = np.linspace(0,1260,8,dtype=int)
        new_ticks = np.linspace(0,210,8,dtype=int)
        plt.xticks(old_ticks,new_ticks)
    else:
        plt.ylim(0,14)
        plt.xlim(0,720)

        old_ticks = np.linspace(0,720,9,dtype=int)
        new_ticks = np.linspace(0,120,9,dtype=int)
        plt.xticks(old_ticks,new_ticks)
    
   

    plt.title(title_str)
    savename = './fig/{}.png'.format(title_str)
    plt.savefig(savename,dpi = 1000)
    
    plt.show()
    
    
```

```python
titls_strs = [
    'ANKER 18w current and voltage curve',
    'UGREEN 18w current and voltage curve',
    'XiaoMi 65w current and voltage curve',
    'Apple 10w current and voltage curve',
    'Apple 5w current and voltage curve'
]
```

## 绘制电压电流曲线

```python
for i in range(len(excel_str)):
    plot_voltage_current(excel_str[i],titls_strs[i])
```

## 绘制功率曲线

```python
curve_strs = [
    'ANKER 18w',
    'UGREEN 18w',
    'XiaoMi 65w',
    'Apple 10w',
    'Apple 5w'
]
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111)
for i, files in enumerate(excel_str):
    raw_data=pd.read_excel('./data/{}'.format(files)).values
    datas = pd.DataFrame(raw_data[:,1:],columns=['voltage','current'])
    datas['power'] = datas['voltage']*datas['current']
    ax.plot(datas.index,datas['power'],label=curve_strs[i])
ax.legend()



ax.set_xlabel('Times (min)')
ax.set_ylabel('Power (w)')
ax.yaxis.grid(True)
ax.xaxis.grid(True)
plt.ylim(0,20)
plt.xlim(0,1260)

old_ticks = np.linspace(0,1260,8,dtype=int)
new_ticks = np.linspace(0,210,8,dtype=int)
plt.xticks(old_ticks,new_ticks)
    
plt.title('Power curve on five chargers')

savename = './fig/{}.png'.format('power curve')
plt.savefig(savename,dpi = 1000)    
```

## 电压稳定性测试

9v 电压区间:10-40 min

5v 电压区间:60-90 min

### 9v 稳定性

```python
V_datas = []
labels = [
    'ANKER 18w',
    'UGREEN 18w',
    'XiaoMi 65w',
]
for i in range(3):
    raw_data=pd.read_excel('./data/{}'.format(excel_str[i])).values
    V_datas.append(raw_data[60:240,1].tolist())
    
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
bplot = ax.boxplot(np.array(V_datas).T,patch_artist=True,vert=True,notch=True,labels = labels)
plt.ylim(8.9,9.3)
ax.set_ylabel('Voltage (V)')
colors = ['lightblue', 'lightblue', 'lightblue']
ax.yaxis.grid(True)
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title('Voltage stability in 9V with different chargers')
savename = './fig/{}.png'.format('Voltage stability in 9V')
plt.savefig(savename,dpi = 1000)    
```

### 5v 稳定性

```python
V_datas = []
labels = [
    'ANKER 18w',
    'UGREEN 18w',
    'XiaoMi 65w',
    'Apple 10w',
    'Apple 5w'
]
for i in range(5):
    raw_data=pd.read_excel('./data/{}'.format(excel_str[i])).values
    V_datas.append(raw_data[360:540,1].tolist())
    
fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
bplot = ax.boxplot(np.array(V_datas).T,patch_artist=True,vert=True,notch=True,labels = labels)
#plt.ylim()

colors = ['lightblue', 'lightblue', 'lightblue','lightblue','lightblue']
ax.yaxis.grid(True)
ax.set_ylabel('Voltage (V)')
for patch, color in zip(bplot['boxes'], colors):
    patch.set_facecolor(color)
    
plt.title('Voltage stability in 5V with different chargers')
savename = './fig/{}.png'.format('Voltage stability in 5V')
plt.savefig(savename,dpi = 1000)    
```

## 充电时间

```python
t_datas = []
labels = [
    'ANKER 18w',
    'UGREEN 18w',
    'XiaoMi 65w',
    'Apple 10w',
    'Apple 5w'
]
for i in range(5):
    raw_data=pd.read_excel('./data/{}'.format(excel_str[i])).values
    t_datas.append(raw_data.shape[0]/6)
    

    
def autolabel(rects):
    
    for rect in rects:
        height = int(rect.get_height())
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

fig = plt.figure(figsize=(8,4))
ax = fig.add_subplot(111)
rects = ax.bar(labels,np.array(t_datas),width=.2,color= 'lightblue',edgecolor = 'k')
ax.yaxis.grid(True)
ax.set_ylabel('Times (min)')

autolabel(rects)
plt.title('Charging time with different chargers')
savename = './fig/{}.png'.format('Charging time')
plt.savefig(savename,dpi = 1000)    
```
