```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df=pd.read_excel('./data/anker18w_iphonexr.xlsx')
data = df.values
```

```python
data01 = pd.DataFrame(data[:,1:],columns=['voltage','current'])
```

```python
data01['power'] = data01['voltage']*data01['current']
```

```python

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df.index,data01['current'],'r',label='current')
plt.ylim(0,2.5)
plt.legend(loc = 'upper left')
ax2 = ax1.twinx()
ax2.plot(df.index,data01['voltage'],'b',label='voltage')

plt.legend(loc = 'upper right')

ax1.set_xlabel('Times')
ax1.set_ylabel('current', color='r')
ax2.set_ylabel('voltage', color='b')
plt.ylim(0,14)
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(df.index,data01['power'],'r',label='power')
```
